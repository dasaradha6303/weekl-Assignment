its digits each raised to the power n equals the number itself. Example: 153 = 1³ + 5³ + 3³.) Print them

```python
armstrong=[]
for num in range(1,10000):
  temp=num
  digitscount=0
  while temp>0:
    digitscount+=1
    temp//=10
  temp=num
  digitsum=0
  while temp>0:
    digitsum+=temp%10**digitscount
    temp//=10
  if digitsum==num:
    armstrong.append(num)
print(armstrong)
print(f"count:{len(armstrong)}")
```

Q44. Manual Title Case. Take a sentence (e.g., "hello world from bengaluru"). Capitalize the first
letter of EACH word, but do NOT use .title() or .capitalize(). Use only .upper() / .lower() and string
indexing. Output: "Hello World From Bengaluru".

```python
def manual_title_case(sentence):
    words = sentence.split(" ")
    title_words = []

    for word in words:
        if word:
            first_letter = word[0].upper()
            remaining_letters = word[1:].lower()
            processed_word = first_letter + remaining_letters
            title_words.append(processed_word)
        else:
            title_words.append("")
    return " ".join(title_words)
input_text = "hello world from bengaluru"
output_text = manual_title_case(input_text)

print(f"Output:{output_text}")
```

```python
def caesar_cipher(message, shift):
    encrypted_chars = []

    for char in message:
        if char.isupper():
          shifted_char = chr((ord(char) - 65 + shift) % 26 + 65)
          encrypted_chars.append(shifted_char)
        elif char.islower():
          shifted_char = chr((ord(char) - 97 + shift) % 26 + 97)
          encrypted_chars.append(shifted_char)
        else:
           encrypted_chars.append(char)

    return "".join(encrypted_chars)

given_text = "Hello, India!"
shift_value = 3
output = caesar_cipher(given_text, shift_value)

print(f"Shift Value: {shift_value}")
print(f"Encrypted text: {output}")
```

Q46. Write a function `calculate_gst(amount, gst_rate=18, gst_inclusive=False)`:
- If gst_inclusive=False: amount is base, add GST on top
- If gst_inclusive=True: amount already includes GST, extract it
- Return a tuple: (base_amount, gst_amount, total_amount)
- All values rounded to 2 decimals
Test cases:
calculate_gst(1000)
→ (1000.00, 180.00, 1180.00)
calculate_gst(1180, gst_inclusive=True) → (1000.00, 180.00, 1180.00)
calculate_gst(500, gst_rate=5)
→ (500.00, 25.00, 525.00)

```python
def calculate_gst(amount, gst_rate=18, gst_inclusive=False):
    if gst_inclusive:
      total_amount=amount
      base_amount = total_amount/(1+(gst_rate/100))
      gst_amount = total_amount-base_amount
    else:
        base_amount=amount
        gst_amount = amount*(gst_rate/100)
    total_amount = base_amount + gst_amount
    return round(base_amount, 2), round(gst_amount, 2), round(total_amount, 2)

print(calculate_gst(1000))
print(calculate_gst(1180, gst_inclusive=True))
print(calculate_gst(500, gst_rate=5))
print(calculate_gst(1000, gst_rate=5, gst_inclusive=True))
```

Q47. Write a RECURSIVE function `power(base, exp)` that computes base raised to exp, WITHOUT
using the `**` operator and WITHOUT using `pow()`. Handle exp = 0 (return 1) and negative exp
(return 1/power(base, -exp)). Test with: power(2, 10), power(5, 0), power(3, -2).

```python
def power(base, exp):
    if exp==0:
        return 1
    elif exp < 0:
        return 1/power(base, -exp)
    else:
      return base*power(base, exp-1)

print(power(2, 10))
print(power(5, 0))
print(power(3, -2))
```
